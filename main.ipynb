```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2

path = 'FaceDataset/'


def image_vector(image):
    
    #print('Original Dimensions : ',img.shape) 
    # Reducing Noise
    img = cv2.fastNlMeansDenoising(image,None,10,7,21)
    
    scale_percent = 10 # percent of original size
    width = int(img.shape[1] * scale_percent / 100)
    height = int(img.shape[0] * scale_percent / 100) 
    dim = (width, height) 
    
    # resize image
    resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA) 
    #print(f" image:{image.shape} img:{img.shape}resized:{resized.shape}")
    
    img_vectorized = resized.reshape(resized.shape[0] * resized.shape[1],1)
    y = img_vectorized / img_vectorized.max(axis=0)
    
    return y

def calculate_yhat(model,y):
    #np.set_printoptions(precision=10)
    model_trans= np.transpose(model)
    #print(y.shape,model.shape,model_trans)
    B1 = np.dot(model_trans,model)
    
    B1_inv = np.linalg.pinv(B1)
    B2 = np.dot(model_trans,y)
    B=np.dot(B1_inv,B2)
    y_hat = np.dot(model,B) 
    return y_hat

def calculate_distance(Xi,y):
    y_hat = calculate_yhat(Xi,y)
    dist = np.linalg.norm(y - y_hat)
    return dist

    

def check_class(dist_list):
    class_dist = []
    for i in range(len(dist_list)):
        class_dist.append(dist_list[i][0])
    class_index= class_dist.index(min(class_dist))
    return dist_list[class_index][1]


 # initializing numpy array
X1,X2,X3,X4,X5,X6,X7,X8,X9,X10,\
X11,X12,X13,X14,X15,X16,X17,X18,X19,X20,\
X21,X22,X23,X24,X25,X26,X27,X28,X29,X30,\
X31,X32,X33,X34,X35,X36,X37,X38,X39,X40 = (np.zeros((56,5)) for i in range(40))

 # Main list Containing all Xi matrices
X = [X1,X2,X3,X4,X5,X6,X7,X8,X9,X10,\
X11,X12,X13,X14,X15,X16,X17,X18,X19,X20,\
X21,X22,X23,X24,X25,X26,X27,X28,X29,X30,\
X31,X32,X33,X34,X35,X36,X37,X38,X39,X40]
for i in range(40):
    for j in range(5):
        img_path=os.path.join(path,f's{i+1}','train',f'{j+6}.pgm')
        img = cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
        normal = image_vector(img)
        X[i][:,j] = normal[0]



#test_img = cv2.imread('FaceDataset/s40/test/1.pgm',IMREAD_UNCHANGED)

accurate = []
inaccurate = []
for i in range(40):
    for j in range(5):
        test_img = cv2.imread(f"FaceDataset/s{i+1}/test/{j+1}.pgm",cv2.IMREAD_GRAYSCALE)
        dist_list = []
        y = image_vector(test_img)
        
        #print('About to print X')
        #print(X)
        for k in range(40):    
            dist = calculate_distance(X[k],y)
            dist_list.append((dist,k+1))
        #print(dist_list)
        result = check_class(dist_list)
        if result==i+1:
            accurate.append(result)
        else:
            inaccurate.append(result)

accuracy = len(accurate)*100/len(inaccurate)
print(accuracy)
          
```
