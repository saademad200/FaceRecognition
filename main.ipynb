```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
path = 'FaceDataset/'


def image_vector(img):
    
    img.thumbnail((64, 64), Image.ANTIALIAS)# resizing test image
    img_data = np.asarray(img)
    img_vectorized = img_data.reshape(img_data.shape[0] * img_data.shape[1],1)
    y = img_vectorized / img_vectorized.max(axis=0)
    return y

def calculate_yhat(model,y):
    model_trans= np.transpose(model)
    B1 = np.dot(model_trans,model)
    B1_inv = np.linalg.inv(B1)
    B2 = np.dot(model_trans,y)
    B=np.dot(B1_inv,B2)
    y_hat = np.dot(model,B) 
    return y_hat

def calculate_distance(Xi,y):
    try:
        y_hat = calculate_yhat(Xi,y)
        dist = np.linalg.norm(y-y_hat)
        return dist
    except:
        pass

def check_class(dist_list):
    temp=dist_list[0][0]
    temp_index=0
    for i in range(1,len(dist_list)):
        
        if dist_list[i][0]:
            if dist_list[i][0] < temp:
                temp = dist_list[i][0]
                temp_index = i
    return temp_index


 # initializing numpy array
X1,X2,X3,X4,X5,X6,X7,X8,X9,X10,\
X11,X12,X13,X14,X15,X16,X17,X18,X19,X20,\
X21,X22,X23,X24,X25,X26,X27,X28,X29,X30,\
X31,X32,X33,X34,X35,X36,X37,X38,X39,X40 = (np.zeros((3392,5)) for i in range(40))

 # Main list Containing all Xi matrices
X = [X1,X2,X3,X4,X5,X6,X7,X8,X9,X10,\
X11,X12,X13,X14,X15,X16,X17,X18,X19,X20,\
X21,X22,X23,X24,X25,X26,X27,X28,X29,X30,\
X31,X32,X33,X34,X35,X36,X37,X38,X39,X40]

for i in range(40):
    for j in range(5):
        img_path=os.path.join(path,f's{i+1}','train',f'{j+6}.pgm')
        img = Image.open(img_path)
        img.thumbnail((64, 64), Image.ANTIALIAS)  # resizes image in-place
        data = np.asarray(img)
        vectorized = data.reshape(data.shape[0] * data.shape[1],1)
        normal = vectorized / vectorized.max(axis=0)
       
        X[i][:,j] = normal[0]
        
test_img = Image.open('FaceDataset/s40/test/1.pgm')

accurate = []
inaccurate = []
for i in range(40):
    for j in range(5):
        test_img = Image.open(f"FaceDataset/s{i+1}/test/{j+1}.pgm")
        dist_list = []
        y = image_vector(test_img)
        for k in range(40):    
            dist = calculate_distance(X[k],y)
            dist_list.append((dist,k+1))
        result = check_class(dist_list)
        if result==i:
            accurate.append(result)
        else:
            inaccurate.append(result)

accuracy = len(accurate)*100/len(inaccurate)
print(accuracy)
          
```
